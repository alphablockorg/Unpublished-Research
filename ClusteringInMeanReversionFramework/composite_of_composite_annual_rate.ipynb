```python
from dependencies import get_polynomial_features

xs, ys, xs_pred_fifteen_years, y_pred_fifteen_years = get_polynomial_features(fifteen_years_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_five_years, y_pred_five_years = get_polynomial_features(five_years_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_quarterly, y_pred_quarterly = get_polynomial_features(quarterly_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_seven_years, y_pred_seven_years = get_polynomial_features(seven_years_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_two_years, y_pred_two_years = get_polynomial_features(two_years_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_weekly, y_pred_weekly = get_polynomial_features(weekly_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_monthly, y_pred_monthly = get_polynomial_features(monthly_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_yearly, y_pred_yearly = get_polynomial_features(yearly_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_ten_years, y_pred_ten_years = get_polynomial_features(ten_years_data_frame, ['Relative_Percentile', 'Annual_Rate'])
xs, ys, xs_pred_three_years, y_pred_three_years = get_polynomial_features(three_years_data_frame, ['Relative_Percentile', 'Annual_Rate'])
```

```python
import numpy as np

regression_array = [
    [xs_pred_fifteen_years, y_pred_fifteen_years, 'fifteen_years', 'blue'],
    [xs_pred_five_years, y_pred_five_years, 'five_years', 'green'],
    [xs_pred_quarterly, y_pred_quarterly, 'quarterly', 'red'],
    [xs_pred_seven_years, y_pred_seven_years, 'seven_years', 'cyan'],
    [xs_pred_two_years, y_pred_two_years, 'two_years', 'magenta'],
    [xs_pred_three_years, y_pred_three_years, 'three_years', 'black'],
    [xs_pred_yearly, y_pred_yearly, 'yearly', 'purple'],
    [xs_pred_ten_years, y_pred_ten_years, 'ten_years', 'orange'],
]
```

```python
index = 0
for arr in regression_array:
    plt.plot(arr[0], arr[1], color=arr[3], label=arr[2])
    index += 1

plt.xlabel('Relative_Percentile')
plt.ylabel('Annual Rate')
plt.legend()
plt.show()
```
